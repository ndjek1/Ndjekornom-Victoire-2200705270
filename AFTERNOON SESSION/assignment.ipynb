```python
# fixing incorrect format in dayte column
df['Date'] = pd.to_datetime(df['Date'],format='mixed')
# Filling null cells 
for x in df.index:
    if pd.isnull(df.loc[x,'Date']):
        df.loc[x,'Date'] = (df.loc[x - 1,'Date'] + pd.Timedelta(days=1))

df
```

```python
# Row 7 of Duration is probably a typo 
df.loc[7,'Duration'] = 40
print(df.to_string())

for x in df.index:
    if df.loc[x,'Duration'] >60:
        df.drop(x, inplace = True)
```

```python
# fill null values in calories column using the mode()
# x = df['Calories'].mode()[0]

# df.fillna({'Calories':x}, inplace = True)
# df
featires = ['Calories', 'Maxpulse', 'Pulse']
for feature in featires:
    x = df[feature].median()
    for i in df.index:
        if pd.isna(df.loc[i,feature]):
            df.loc[i, feature] = x

df
```

```python
# fill null values in Maxpulse column using median()
x = df['Maxpulse'].median()

df.fillna({'Maxpulse':x}, inplace = True)
df
```

```python
# fixing incorrect format in date column
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed')
df
```

```python
# second row has too much null values and we cannot fix that so we just drop it

df.drop([2],inplace=True)
df
```

```python
features1 = ['Quantity', 'Total Revenue']

for feature in features1:
    for i in df.index:
        if df.loc[i, feature] <0:
            df.loc[i, feature] = abs(df.loc[i, feature])

df
```

```python
# Unit price 


for i in df.index:
    if pd.isna(df.loc[i, 'Unit Price'] ):
        df.loc[i, 'Unit Price'] = df.loc[i, 'Total Revenue']/df.loc[i, 'Quantity']
df
```
